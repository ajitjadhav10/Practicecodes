Heatmap : Computing the correlation matrix

- Correlation is a term used to represent the statistical measure of linear relationship between two variables. It can also be defined as the measure of dependence between two different variables.

- Region wise trend
- Latin America & Caribbean','South Asia','East Asia & Pacific','Europe & Central Asia','Middle East & North Africa','North America','Sub-Saharan Africa'

```python
LAC=new_data_2.groupby('Region').get_group('Latin America & Caribbean')
LAC.head(3)
```

```python
LAC.head(3)
```

```python
LAC.loc[:,"Country Name"].head(2)
```

```python
plt.figure(figsize=(25,8))
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"1990"],color='red',label='Year: 1990')
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"2010"],color='green',label='Year: 2010')
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"2019"],color='blue',label='Year: 2019') 
plt.xticks(rotation=90)
plt.grid(linewidth=0.4)
plt.title("Percentage of women in workforce vs Latin America & Caribbean countries for the year")
plt.xlabel("Latin America and Caribbean Countries")
plt.ylabel("Percentage of Women in Workforce")
plt.legend()
```

```python
SA=new_data_2.groupby('Region').get_group('South Asia')
SA.head(3)
```

```python
plt.figure(figsize=(25,8))
plt.plot(SA.loc[:,"Country Name"],SA.loc[:,"1990"],color='red',label='Year: 1990')
plt.plot(SA.loc[:,"Country Name"],SA.loc[:,"2010"],color='green',label='Year: 2010')
plt.plot(SA.loc[:,"Country Name"],SA.loc[:,"2019"],color='blue',label='Year: 2019') 
plt.xticks(rotation=90)
plt.grid(linewidth=0.4)
plt.title("Percentage of women in workforce vs Latin America & Caribbean countries for the year")
plt.xlabel("Latin America and Caribbean Countries")
plt.ylabel("Percentage of Women in Workforce")
plt.legend()
```
