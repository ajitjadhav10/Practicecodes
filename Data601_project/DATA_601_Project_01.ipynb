- Histogram of how percentage of women in worforce has changed region wise between 1990 to 2019.

```python
df_1 = new_data_2[['1990','2019','Region']]

regions = ['Latin America & Caribbean','South Asia','East Asia & Pacific','Europe & Central Asia','Middle East & North Africa','North America','Sub-Saharan Africa']
df_1 = df_1[ df_1['Region'].isin(regions)  ]

df_1
```

```python
g = sns.FacetGrid(df_1, col="Region")
g.map(plt.hist, "1990")
g = sns.FacetGrid(df_1, col="Region")
g.map(plt.hist, "2019")
```

- Analyzing the countries within each region

```python
#1. Having a look at countries within Latin America & Caribbean region

LAC=new_data_2.groupby('Region').get_group('Latin America & Caribbean')
LAC.head(3)
```

```python
LAC.head(3)
```

```python
LAC.loc[:,"Country Name"].head(2)
```

```python
#Country wise trend of women in workforce in Latin America & Caribbean region

plt.figure(figsize=(25,8))
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"1990"],color='red',label='Year: 1990')
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"2010"],color='green',label='Year: 2010')
plt.plot(LAC.loc[:,"Country Name"],LAC.loc[:,"2019"],color='blue',label='Year: 2019') 
plt.xticks(rotation=90)
plt.grid(linewidth=0.4)
plt.title("Percentage of women in workforce vs Latin America & Caribbean countries for the year")
plt.xlabel("Latin America and Caribbean Countries")
plt.ylabel("Percentage of Women in workforce")


plt.legend()
```
